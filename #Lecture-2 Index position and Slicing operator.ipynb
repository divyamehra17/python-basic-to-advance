```python
# step size -
# syntax [start:end:step size]
```

```python
#integer
a=2456256
a[0:4:1]
```

```python
#float
b=45645.2365
b[0:4:1]
```

```python
#complex
c=45+6j
c[0:4:1]
```

```python
#boolean
d=False
d[0:4:1]
```

```python
#string
e='i am working as a teacher'
print(e)
```

```python
# start point and end point both are positive value and step size = 1,start point is less value and end point is more value
e[0:5:1]     
```

```python
# start point and end point both are positive value and step size = -1,start point is less value and end point is more value
e[0:5:-1]  
```

```python
# start point and end point both are positive value and step size = 1,start point is more value and end point is less value
e[5:1:1]
```

```python
# start point and end point both are positive value and step size = -1,start point is more value and end point is less value(Reverse order)
e[5:1:-1]
```

```python
# start point and end point both are negative value and step size = 1,start point is less value and end point is more value
e[-1:-5:1]
```

```python
# start point and end point both are negative value and step size = -1,start point is less value and end point is more value (Reverse order)
e[-1:-5:-1]
```

```python
# start point and end point both are negative value and step size = 1,start point is more value and end point is less value
e[-5:-1:1]
```

```python
# start point and end point both are negative value and step size = -1,start point is more value and end point is less value
e[-5:-1:-1]
```

```python
# start point is negative value and end point is positive value  and step size = 1,start point is less value and end point is more value
e[-1:5:1]
```

```python
# start point is negative value and end point is positive value and step size = -1,start point is less value and end point is more value(Reverse order)
e[-1:5:-1]
```

```python
# start point is positive value and end point is negative value and step size = 1,start point is more value and end point is less value
e[1:-5:1]
```

```python
# start point is positive value and end point is negative value and step size = -1,start point is more value and end point is less value
e[1:-5:-1]
```
